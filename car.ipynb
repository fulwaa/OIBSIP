```python
# importing libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error
```

```python
#read the file
data = pd.read_csv("H:\\oasis_internship\\car\\CarPrice_Assignment.csv")
```

```python
#checking uniqueness of each column
data.nunique()
```

```python
icol=['fueltype','aspiration','doornumber', 'carbody', 'drivewheel', 'enginelocation','enginetype',
       'cylindernumber','fuelsystem']
```

```python
#converting to categorical
ca=preprocessing.LabelEncoder()
for i in range(len(icol)):
    data[icol[i]]=ca.fit_transform(data[icol[i]])
```

```python
data.head()
```

```python
x=data['price']
```

```python
col=['car_ID', 'symboling', 'fueltype', 'aspiration',
       'doornumber', 'carbody', 'drivewheel', 'enginelocation', 'wheelbase',
       'carlength', 'carwidth', 'carheight', 'curbweight', 'enginetype',
       'cylindernumber', 'enginesize', 'fuelsystem', 'boreratio', 'stroke',
       'compressionratio', 'horsepower', 'peakrpm', 'citympg', 'highwaympg']
```

```python
y=data[col]
```

```python
#splitting
X_train, X_test, y_train, y_test = train_test_split(y, x, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
#model building
model = RandomForestRegressor()
```

```python
#model training
model.fit( X_train,y_train)
```

```python
#model prediction
pred=model.predict(X_test)
```

```python
#checking mean absolute error
mean_absolute_error(y_test,pred)
```
